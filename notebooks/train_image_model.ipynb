# Train Image Recognition Model (Custom CNN)

This notebook trains a custom CNN on the mock image data (or Food-101 if provided).

```python
import os
import sys
import tensorflow as tf
import matplotlib.pyplot as plt

# Add root to path
sys.path.append(os.path.abspath('..'))

from models.image_classifier import build_custom_cnn, load_data
```

```python
data_dir = '../data/mock_images'
# Check if real data exists, else use mock
if os.path.exists('../data/food-101/images'):
    data_dir = '../data/food-101/images'
    print("Using Real Food-101 Data")
else:
    print("Using Mock Data")

train_ds, val_ds = load_data(data_dir)
class_names = train_ds.class_names
print("Classes:", class_names)
```

```python
num_classes = len(class_names)
model = build_custom_cnn(num_classes)
model.summary()
```

```python
epochs = 5
history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
```

```python
# Save model
model_save_path = '../models/saved/image_model.keras'
model.save(model_save_path)
print(f"Model saved to {model_save_path}")
```
